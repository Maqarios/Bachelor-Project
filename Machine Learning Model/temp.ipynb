```python
# Function To Calculate PCA
def calculate_PCA(signals, n_components):
    
    signals_temp = numpy.zeros((signals.shape[0], signals.shape[1], n_components))

    pca = PCA(n_components)
    for epoch in range(signals.shape[0]):
        #signals_temp[epoch] = normalize(signals[epoch])
        signals_temp[epoch] = pca.fit_transform(signals[epoch])
    
    return signals_temp

n_components = 4

# A
Subject_A_Signal_Train_PCA = numpy.zeros((Subject_A_Signal_Train.shape[0], Subject_A_Signal_Train.shape[1], EXTRACTED_CHANNELS.shape[0]))
Subject_A_Signal_Test_PCA = numpy.zeros((Subject_A_Signal_Test.shape[0], Subject_A_Signal_Train.shape[1], EXTRACTED_CHANNELS.shape[0]))
for channel_index in range(EXTRACTED_CHANNELS.shape[0]):
    Subject_A_Signal_Train_PCA[:, :, channel_index] = Subject_A_Signal_Train[:, :, EXTRACTED_CHANNELS[channel_index]]
    Subject_A_Signal_Test_PCA[:, :, channel_index] = Subject_A_Signal_Test[:, :, EXTRACTED_CHANNELS[channel_index]]

Subject_A_Signal_Train_PCA = calculate_PCA(Subject_A_Signal_Train_PCA, n_components)
Subject_A_Signal_Test_PCA = calculate_PCA(Subject_A_Signal_Test_PCA, n_components)

Subject_A_Train_Prediction_PCA = predict_characters(
    Subject_A_Signal_Train_PCA,
    Subject_A_StimulusCode_Train,
    numpy.arange(n_components),
    Subject_A_Weights_PCA,
    MATRIX,
    start_window=START_WINDOW,
    end_window=END_WINDOW
)

Subject_A_Test_Prediction_PCA = predict_characters(
    Subject_A_Signal_Test_PCA,
    Subject_A_StimulusCode_Test,
    numpy.arange(n_components),
    Subject_A_Weights_PCA,
    MATRIX,
    start_window=START_WINDOW,
    end_window=END_WINDOW
)

print('A Train Prediction PCA:', calculate_accuracy(Subject_A_TargetChar_Train[0], Subject_A_Train_Prediction_PCA))
print('A Test Prediction PCA:', calculate_accuracy(SUBJECT_A_TEST_RESULTS, Subject_A_Test_Prediction_PCA))
```
